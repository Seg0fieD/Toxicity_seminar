```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os 
```

```python
result_path = "../results/metrics.csv"
if not os.path.exists(result_path):
    print("Create metrics.csv file in the results folder")

result_db = pd.read_csv(result_path)
```

### Eye Irritation

```python
eye_irritation = result_db[result_db["model_name"].str.contains("Eye_Irritaion", case= False)]
eye_irritation
```

```python
colmn = eye_irritation.columns
model_names = eye_irritation[colmn[0]].values


values = {}
for i in range(1, len(colmn)):
    values[colmn[i]] = eye_irritation[colmn[i]].values
    

x = np.arange(len(model_names))  
width = 0.25  
multiplier = 0

fig, ax = plt.subplots(figsize= (10, 8) )#layout='constrained')

for attribute, measurement in values.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3, rotation= 80)
    multiplier += 1


ax.set_title('Whole Validation Set for Eye Irritation')
ax.set_xticks(x + width, labels= model_names, rotation= 45, ha= "right")
ax.set_yticks(np.arange(0, 1.2, 0.1))
ax.legend(loc='upper left', ncols=3)

plt.show()
fig.savefig(fname= "../results/Eye_Irritation.png", bbox_inches = "tight") 
```
